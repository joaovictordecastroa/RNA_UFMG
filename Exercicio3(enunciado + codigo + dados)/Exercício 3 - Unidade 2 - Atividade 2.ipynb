```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
import random
import os
import sys
module_path = os.path.abspath(os.path.join('./'))
if module_path not in sys.path:
    sys.path.append(module_path)
from Adaline import Adaline


inputs = pd.read_csv('./x.csv', sep=" ")
Y = pd.read_csv('./y.csv', sep=" ")
T = pd.read_csv('./t.csv', sep=" ")

fig = go.Figure(data=go.Scatter(x=T.t, y=inputs.V1, mode='lines+markers', name='X1'))
fig.add_trace(go.Scatter(x=T.t, y=Y["V1"], mode='lines+markers', name='Y'))
fig.add_trace(go.Scatter(x=T.t, y=inputs["V2"], mode='lines+markers', name='X2'))
fig.add_trace(go.Scatter(x=T.t, y=inputs["V3"], mode='lines+markers', name='X3'))
fig.update_layout(title='Entradas e Saídas', width=1024, height=516)


fig.show()


    
```

```python
adaline = Adaline(3)

train_size = int(inputs.shape[0] * 0.7)
test_size = int(inputs.shape[0] * 0.3)

dataframe = [inputs.V1, inputs.V2, inputs.V3]
adaline.train(inputs[['V1', 'V2', 'V3']].values, Y.V1)

print('Pesos encontrados: ',adaline.weights)
```

```python
test_result = adaline.test(inputs[['V1', 'V2', 'V3']].values[:test_size], Y.V1[:test_size])

print('Erro quadrático médio: ', test_result['mean_square_error'])

erros = go.Figure(go.Scatter(x=np.arange(1, len(adaline.errors_per_epoch)), y=(np.array(adaline.errors_per_epoch)), mode='lines', name='Erro por década'))
erros.show()
```
