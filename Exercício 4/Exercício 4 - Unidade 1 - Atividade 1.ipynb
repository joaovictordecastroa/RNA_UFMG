```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
import random
import os
import sys
from Perceptron import Perceptron

s1 = 0.4
s2 = 0.4
nc = 200

xc1 = [np.random.normal(size = nc*2, loc = 2, scale= s1), np.random.normal(size = nc*2, loc = 2, scale= s1)]
xc2 = [np.random.normal(size = nc*2, loc = 4, scale= s2), np.random.normal(size = nc*2, loc = 4, scale= s2)]

data = {
    'x': np.concatenate((xc1[0], xc2[0])),
    'y':  np.concatenate((xc1[1], xc2[1]))
}

df = pd.DataFrame(data=data)



inputs = (df.values.tolist())

outputs = np.concatenate((np.zeros(nc*2), np.ones(nc*2)))

```

```python
fig = go.Figure(data=go.Scatter(x=xc1[0], y=xc1[1], mode='markers', name='xc_1', marker=dict(
        size=8,
        line=dict(width=2,
        color='DarkSlateGrey'),
        color='#FD3216'
    )))

fig.add_trace(go.Scatter(x=xc2[0], y=xc2[1], mode='markers', name='xc_2', marker=dict(
        size=8,
        line=dict(width=1,
        color='DarkSlateGrey'),
        color='#6A76FC'
    )))


fig.show()
```

```python
treina = Perceptron(2)
result = treina.train(inputs, outputs)
print('Pesos: ', result)
print('Erro:', treina.erro)

# Calculate the Decision Boundary of a Single Perceptron - Visualizing Linear Separability


# The equation of a line is defined as Ax + By - C = 0

# To calculate linear separability, we will use the weights to use as the parameters of the straight

# In this case, A is W2, B is W1 and C is the Bias

A = result[2]
B = result[1]
C = result[0]

# equation of straight:
print('A equação da reta: {A}x + {B}y + {C} = 0'.format(A=A, B=B, C=C))
# Now, we calculate the x axis interception on the y axis
x_intercection = 0 if C == 0 else -C / A
y_intercection = 0 if C == 0 else -C / B 

# now, we calculate the slope of the equation
m =  - (y_intercection / x_intercection)


# Define the straight equation
def straight_function(x):
    axis = []
    for i in range(len(x)):
        result = (m * x[i] + y_intercection)
        axis.insert(i,result)
    return axis


x = np.linspace(0, 6, num=nc)
y = straight_function(x) 

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='xc_2', marker=dict(
         size=8,
         line=dict(width=1,
         color='DarkSlateGrey'),
         color='#6A76FC'
     )))
fig.show()
```

```python
print('Superfície de separação')

x = np.linspace(0, 6, num=nc)
y = np.linspace(0, 6, num=nc)

z = []

for i in range(nc):
    point_x = x[i]
    point_y = y[i]
    
    point_on_straight = A*point_x + B*point_y + C
    if (point_on_straight >= 0):
        z.insert(i,1)
    else:
        z.insert(i,0)
z = [z]*nc    
fig3d = go.Figure(data=[go.Surface(x=x, y=y, z=z)])
fig3d.show()
```
