```python
import math
import itertools
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go


class circle:
    @staticmethod
    def hypotenuse(x, y):
        return math.sqrt(math.pow(x,2) + math.pow(y,2))   

# initialize a list with a range of -1 to 1 where the values ​​are evenly spaced within a given range of 0.1.x = np.arange(-1, 1, 0.1)
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)

colors = []
z = []

# initializing a grid matrix with the given points x and y
create_grid = itertools.product(x,y)

# initializing a panda dataframe with grid matrix
dataframe = pd.DataFrame(create_grid, columns=['x','y'])

raio = 0.6

for index, row in dataframe.iterrows():
    x = row['x']
    y = row['y']

    # Checks whether the x and y points are within the radius of 0.6
    if (circle.hypotenuse(x,y) > raio):
        # set value with cordinate x and y with red color to represents the class Red
        colors.append("red")
        z.append(-1)
    else:
        # set value with cordinate x and y with red color to represents the class Black
        colors.append("black")
        z.append(1)

# create a new dataframe column with colors list and z values
dataframe['colors'] = colors
dataframe['z'] = z

# generate a 2d plot with a dataframe grid values representing the two classes
fig = px.scatter(dataframe, x=dataframe['x'], y=dataframe['y'], color=dataframe['colors'], color_discrete_sequence=['red', 'black'], template="seaborn", width=800, height = 800)

fig.update_traces(hoverinfo='skip', marker=dict(size=12,
                              line=dict(width=2,
                                        color='black')),
                  selector=dict(mode='markers'))
#plot the graphic
fig.show()

fig = px.scatter_3d(dataframe, x=dataframe['x'], y=dataframe['y'], z=dataframe['z'], color=dataframe['colors'], color_discrete_sequence=['red', 'black'], template="seaborn",width=1200, height=800)

fig.update_xaxes(showline=True, linewidth=1, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=1, linecolor='black', mirror=True)

fig.update_traces(hoverinfo='skip', marker=dict(size=8,
                              line=dict(width=2,
                                        color='black')),
                  selector=dict(mode='markers'))

# z_index = np.full((dataframe.shape[0], 1), 0)

# yy, xx = np.meshgrid(np.linspace(-1,1), np.linspace(-1,1))
# zz = yy*0



# fig1 = go.Surface(z=zz, x=xx, y=yy,  colorscale=[[1, 'rgb(250,250,250)']], showscale=False)

fig.update_layout(template="plotly", font_size=10)
fig.update_layout(scene_aspectmode='auto')
fig.update_layout(scene = dict(
                    xaxis = dict(
                         showticklabels =False,
                         linecolor='black',
                         linewidth=2,
                         showgrid=False,
                         showspikes=False,
                         showaxeslabels=False

                         ),
                    yaxis = dict(
                        showticklabels =False,
                        linecolor='black',
                        showgrid=False
                        ),
                    zaxis = dict(
                        showticklabels =False,
                        linecolor='black',
                        showgrid=False
                        ),),
                    width=700,
                    plot_bgcolor='black',
                    margin=dict(
                    r=10, l=10,
                    b=10, t=10),
                    coloraxis=dict(colorbar=dict(bordercolor='black'))
                  )

# fig.add_trace(fig1)


fig.show()
# fig.write_html("./plot.html")
```
